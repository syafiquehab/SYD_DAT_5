```python
all_trains = pd.read_csv('data/sydtrains.csv', thousands=",", index_col=1)
all_trains
```

```python
cbd2014 = all_trains[(all_trains.LINE == 'CBD') & (all_trains.YEAR == 2014)]
```

```python
cbd2014
```

```python
cbd2014.IN_24_HOURS.plot.bar()
```

```python
cbd2014.IN_24_HOURS.plot.bar(color=['blue','blue','blue','blue','red','blue','red','blue'])
```

```python
cbd2014.index
```

```python
colour_scheme = [('red' if x in ['Circular Quay', 'Museum'] else 'blue') for x in cbd2014.index]
```

```python
colour_scheme
```

```python
cbd2014.IN_24_HOURS.plot.bar(color=colour_scheme)
```

```python
import numpy as np
```

```python
np.where(cbd2014.IN_0200_0600 < 400,'blue','red')
```

```python
cbd2014.IN_24_HOURS.plot.bar(color=np.where(cbd2014.IN_0200_0600 < 400,'blue','red'))
```

```python
matplotlib.pyplot.savefig('trains.png')
```

```python
morning_chart = cbd2014.plot.scatter(x='IN_0600_0930', y='OUT_0600_0930', color='red', s=[200,150,100,5,5,5,5,5])
cbd2014.plot.scatter(x='IN_1500_1830', y='OUT_1500_1830', color='blue', s=[200,150,100,5,5,5,5,5], ax=morning_chart)
matplotlib.pyplot.xlabel("Passengers In")
matplotlib.pyplot.savefig("train-chart.jpg")
```
